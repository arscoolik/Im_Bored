```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import iqr
%matplotlib inline
```

```python
val = [3, 10]
for n in val:
  ans = 0
  for i in range(1, n + 1):
    ans += i * np.exp(i)
  print(ans)
```

Problem 2

```python
x = np.arange(-10, 10, 0.1)
y = x * x * np.sin(2*x)
```

```python

plt.figure(figsize=(10,6))

plt.scatter(x, y, label = "label_name" )

# Set x and y axes labels
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.title('f(x)')
plt.legend()
plt.show()
```
